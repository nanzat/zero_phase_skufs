```python
import pandas as pd
```

## Загружаем датасет

```python
data = pd.read_csv('../data/_data.csv')
```

```python
data.head()
```

## Удалим столбцы явно не имеющие связи с ценой

```python
data.drop(columns=['Unnamed: 0', 'Телефоны', 'Описание', 'Ссылка на объявление', 'Тип'],inplace=True)
data.head()
```

## Просмотрим количество уникальных значений в столбцах

```python
for i in data.columns:
    print(f"{i}: {len(data[i].unique())}")
```

## Удалим столбцы в которых много уникальных не числовых значений

```python
data.drop(columns=['Серия дома', 'Дополнительно', 'Название ЖК', 'Серия дома', 'Дом'],inplace=True)
data.head()
```

## Смотрим количество пустых ячеек по колонкам

```python
print("Количество пропусков в каждом столбце:")
data.isnull().sum()
```

## Далее пройдемся по столбцам и заполним пропуски

## "Количество комнат" сразу преобразуем в числа и заполним пропуски модой

```python
data['Количество комнат'] = data['Количество комнат'].apply(lambda x: int(x.split(",")[0]) if type(x) is str else x)
data['Количество комнат'] = pd.to_numeric(data['Количество комнат'], downcast='integer')
```

```python
data['Количество комнат'].fillna(int(data['Количество комнат'].mode()), inplace=True)
data.isnull().sum()
```

## 

## В "Метро" оставим только название станций и удалим строки с пропусками

```python
data = data[data['Метро'].notna()]
data['Метро'] = data['Метро'].apply(lambda x: x.split()[1])
data['Метро'].nunique()
```

## В "Адрес" оставим название города

```python
data['Адрес'] = data['Адрес'].apply(lambda x: x.split(", ")[0])
data['Адрес'].nunique()
```

## В "Парковка" нулевые значения заменим на "нет"

```python
data['Парковка'].fillna(value='нет', inplace=True)
data.isnull().sum()
```

## В "Ремонт" заменим пустые ячейки на "Без ремонта"

```python
data['Ремонт'].fillna(value='Без ремонта', inplace=True)
data.isnull().sum()
```

## "Площадь комнат, м2" удалим, поскольку в "Площадь, м2" есть сумма данных значений

```python
data.drop(columns='Площадь комнат, м2', inplace=True)
data.columns
```

## В "Балкон" заменим пустые ячейки на "нет"

```python
data['Балкон'].fillna(value='нет', inplace=True)
data.isnull().sum()
```

## Удалим колонки, которые имеют много пропусков

```python
data.drop(columns=['Окна','Санузел', 'Высота потолков, м', 'Можно с детьми/животными', 'Лифт', "Мусоропровод"], inplace=True)
data.columns
```

## Переименуем столбцы на английский

```python
data.rename(columns={
    'ID  объявления': 'id',
    'Количество комнат': 'num_rooms',
    'Метро': 'metro',
    'Адрес': 'address',
    'Площадь, м2': 'area_m2',
    'Парковка': 'parking',
    'Цена': 'price',
    'Ремонт': 'repair',
    'Балкон': 'balcony',
}, inplace=True)
```

```python
data.columns
```

## Сохраним полученный датасет

```python
data.to_csv('data.csv', index=False)
```
